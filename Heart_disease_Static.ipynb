```python
def colum_mean_calculator(input_dist, sample_size, n_samples):
    sample_means = []
    for i in range(n_samples):
        # Take a random sample of size 'sample_size' from 'input_dist' with replacement
        sample = input_dist.sample(sample_size, replace=True)
        # Calculate the mean of this sample and add to the list
        sample_means.append(sample.mean())
    return sample_means
```
